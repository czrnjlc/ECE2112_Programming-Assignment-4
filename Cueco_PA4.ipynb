# ECE Board Exam Problem
Using data wrangling and visualization techniques with storytelling, analyze the dataset provided and present the following:

**1. Data Frames**
* **Instru DataFrane:** Filtered where the track is 'Instrumentation', hometown is 'Luzon', and the 'Electronics' grade is greater than 70. The output includes column 'Name', 'GAES', and 'Electronics'.
*  **Mindy DataFrame:** Filtered where homwtown is 'Mindanao', gender is 'Female', and 'Average' is greater than or equal to 55. The output includes columns 'Name', 'Track, 'Electronics', and 'Average'.

**2. Visualization**
Visualize how different features such as chosen track in college, gender, or hometown contribute to a higher average grade

```python
# Importing the pandas library
import pandas as pd
```

### Task 1: Load Dataset

```python
# Load the dataset
df = pd.read_excel('board2.xlsx')
# Display the dataset
df
```

### Task 2: Filter Data for Intru
Where track is constant as Intrumentation and hometown Luzon.

#### Code Implementation:

```python
# Filter where Track is Instrumentation, Hometown is Luzon, and Electronics grade > 70
Instru = df[(df["Track"] == "Instrumentation") & (df["Hometown"] == "Luzon") & (df["Electronics"] > 70)][["Name", "GEAS", "Electronics"]]
# Display the filtered DataFrame
Instru
```

#### Expected Outcome:
Contains the names, GEAS grades, and Electronics grades of board takers from the Instrumentation track in Luzon, with Electronics grade greater than 70.

### Task 3: Filter Data for Mindy
Where hometown is constant as Mindanao and gender is Female.

#### Code Implementation:

```python
# Calculate the average grade for each student by taking the mean of the "Math", "Electronics",
# "GEAS", and "Communication" columns. The mean is calculated along axis=1
df['Average'] = df[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)
# Filter where Hometown is Mindanao, Gender is Female, and Average >= 55
Mindy = df[(df["Hometown"] == "Mindanao") & (df["Gender"] == "Female") & (df["Average"] >= 55)][["Name", "Track", "Electronics", "Average"]]
# Display the filtered DataFrame
Mindy
```

#### Expected Outcome:
Contains the names, track, Electronics grades, and average grades of female students from Mindanao with an average score of 55 or higher.

### Task 4: Create Visualization
Create a visualization that shows how the different features contribute to average grades. Does chosen track in college, gender, or hometown contributes to a higher average score?

#### Code Implementation:

```python
#import seaborn and matplotlib.pyplot's library
import seaborn as sns
import matplotlib.pyplot as plt

#Set figure size
plt.figure(figsize=(5, 10))

# Bar plot for Track
plt.subplot(3, 1, 1)
sns.barplot(x='Track', y='Average', data=df)  
plt.title('Average Grade by Track')

# Bar plot for Gender
plt.subplot(3, 1, 2)
sns.barplot(x='Gender', y='Average', data=df)  
plt.title('Average Grade by Gender')

# Bar plot for Hometown
plt.subplot(3, 1, 3)
sns.barplot(x='Hometown', y='Average', data=df) 
plt.title('Average Grade by Hometown')

#Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

#### Expected Outcome:
The bar plots show how the selected features contribute to the average grade.
